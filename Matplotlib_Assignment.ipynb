```python
""" Q.2
A scatter plot is a type of data visualization that uses dots (or markers) on a two-dimensional plane to represent the values obtained for two different variables. Each dot on the scatter plot represents a single data point with its respective values for the two variables being plotted.
Scatter plots are particularly useful for visualizing the relationship between two continuous variables and identifying patterns or trends in the data.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")  # creating xlabel as x axis
plt.ylabel("y axis")  # creating ylabel as yaxis
plt.title("random data") #creating title as random data
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,1)
plt.plot(x1,y1,color='orange')
plt.subplot(2,2,2)
plt.plot(x2,y2,color='blue')
plt.subplot(2,2,3)
plt.plot(x3,y3,color='green')
plt.subplot(2,2,4)
plt.plot(x4,y4,color='red')
plt.show()
```

```python
""" Q.4
A bar plot also known as a bar chart or bar graph, is a type of data visualization that uses rectangular bars to represent data values. Each bar's length or height corresponds to the magnitude of a particular variable, allowing viewers to easily compare different data points.
Bar plots are commonly used to display and compare the frequencies, counts, or values of different categories or groups within a dataset.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)  #a bar plot
```

```python
plt.barh(company,profit)  # a horizontal bar plot
```
